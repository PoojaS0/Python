```python
#palindrome............eg=121,222,212..etc

num=int(input("enter the number"))
original=num
reverse=0
while(num>0):
    rem=num%10
    reverse=reverse*10+rem
    num=num//10
print(original,reverse)
if(original==reverse):
    
    print("it is a palindrome")
else:
    print("not palindrome number")
```

```python
# using nested while loop



i=1
while(i<=4):
    j=5
    while(j>=i):
        print(j,end=" ")
        j=j-1
    print()
    i=i+1

#second half
i=5
while(i>=1):
    j=5
    while(j>=i):
        print(j,end=" ")
        j-=1
    print()
    i-=1
        
```

```python
"""
    *
   ***
  *****
 *******
********* """


i=1
while(i<=5):# number of rows needed
    #calculating space
    j=5
    while(j>i):
        print("-",end="")
        j-=1

    #printing star
    k=1
    while(k<=i):
        print("*",end=" ")
        k+=1
    print()
    i+=1
```

```python
i = 5
while(i >= 1):        # start from 5 down to 1

    # printing dashes
    j = 5
    while(j > i):
        print("-", end="")
        j -= 1

    # printing stars
    k = 1
    while(k <= i):
        print("*", end=" ")
        k += 1

    print()
    i -= 1
```
