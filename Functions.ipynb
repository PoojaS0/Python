```python
#write a program to print prime numbers b/w 1 to 10
#1)with i/p without return
#2)with i/p with return
```

```python
#1)with i/p without return


def prime(p1,p2):
    for i in range (p1,p2+1,1):
        for j in range(2,i,1):
            if(i==1):
                print(i)
            elif(i%j==0):
                break
        else:
            print(i)
    
```

```python
prime(20,50)
```

```python
n1=int(input())
n2=int(input())
prime(n1,n2)
```

```python
#2)with i/p with return


def prime1(p1,p2):
    p1=[]
    for i in range (p1,p2+1,1):
        for j in range(2,i,1):
            if(i==1):
                p1.append(i)
            elif(i%j==0):
                break
        else:
             p1.append(i)
    return p1
```

```python
prime1(1,10)
```
